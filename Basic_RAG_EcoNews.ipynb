```python
from langchain_community.document_loaders import CSVLoader

# Directly load CSV
loader = CSVLoader(
    file_path="data/AKSHARE_cjzc/stock_cjzc_em.csv",
    encoding='utf-8-sig'
)
documents = loader.load()
```

```python
documents
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
def split_text(text):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,  # Adjust as needed
        chunk_overlap=100  # Adjust as needed
    )
    return splitter.split_text(text)
```

```python
csv_chunks = []

for doc in documents:
    # chunk every row of the content
    chunks = split_text(doc.page_content)
    # add the chunks to a list
    csv_chunks.extend(chunks)

# print the number of the chunks
print(f"总共切分出 {len(csv_chunks)} 个片段")
```

```python
csv_chunks
```

```python
from langchain_huggingface import HuggingFaceEndpointEmbeddings

model_name = 'intfloat/multilingual-e5-large'

embed = HuggingFaceEndpointEmbeddings(
    model=model_name,
    task="feature-extraction",
    huggingfacehub_api_token=os.environ.get("HUGGINGFACEHUB_API_TOKEN"),
)
```

```python
from langchain_community.vectorstores import FAISS
import faiss

# Embed the csv content and Store in FAISS
def embed_and_store(csv_chunks, embedding_model, index_name):
    faiss_index = FAISS.from_texts(csv_chunks, embedding_model)
    faiss_index.save_local(index_name)
    print(f"FAISS index saved at {index_name}")
    return faiss_index
```

```python
faiss_index = embed_and_store(csv_chunks, embed, "data/cjzc_faiss_index")
```

```python
import numpy as np

def retrieval(query):
    query_embedding = embed.embed_query(query)
    query_embedding_np = np.array(query_embedding).astype("float32").reshape(1, -1)  # Convert to numpy array
    D, I = faiss_index.index.search(query_embedding_np, k=3) # Retrive the top three similar result
    return [csv_chunks[i] for i in I[0]] # Return only the textual content
```

```python
query = "Tell me something about the news in June?"
retrieved_docs = retrieval(query)
for doc in retrieved_docs:
    print(doc)
```

```python
def augment_prompt(query: str):
    # get top 3 results from knowledge base
    results = retrieval(query)
    # get the text from the results
    source_knowledge = "\n".join(results)
    # feed into an augmented prompt
    augmented_prompt = f"""Using the contexts below, answer the query. If the context does not contain enough information for you to answer the query, state it and you might provide general answer based on your knowledge.

    Contexts:
    {source_knowledge}

    Query: {query}"""
    return augmented_prompt
```

```python
print(augment_prompt(query))
```
